```

import os
import glob
import imghdr
import openpyxl
import cv2

INPUT_IMG_DIR = './'
SHEET_TITLE = 'カタログ' # シート名の設定
RESULT_FILE_NAME = 'result.xlsx' # 結果を保存するファイル名


def attach_img(data, ws):
    cnt = 1
    max_width_1 = 0
    max_width_2 = 0
    for k,v in data.items():
        max_height = 0
        ws.cell(row=cnt, column=2).value = k
        img_1 = openpyxl.drawing.image.Image(v[0])
        img_2 = openpyxl.drawing.image.Image(v[1])
        size_img_1 = cv2.imread(v[0])
        size_img_2 = cv2.imread(v[1])
        height_1, width_1 = size_img_1.shape[:2]
        height_2, width_2 = size_img_2.shape[:2]
        if max_width_1 < width_1:
            max_width_1 = width_1
        if max_width_2 < width_2:
            max_width_2 = width_2
        if height_1 < height_2:
            max_height = height_2
        else:
            max_height = height_1
        ws.row_dimensions[cnt].height = max_height * 0.75
        ws.column_dimensions['C'].width = max_width_1 * 0.15
        ws.column_dimensions['D'].width = max_width_2 * 0.15

        cell_address = ws.cell(row=cnt, column=3).coordinate # セルの行列番号から、そのセルの番地を取得
        img_1.anchor = cell_address
        ws.add_image(img_1) # シートに画像貼り付け
        cell_address = ws.cell(row=cnt, column=4).coordinate # セルの行列番号から、そのセルの番地を取得
        img_2.anchor = cell_address
        ws.add_image(img_2) # シートに画像貼り付け




    """
    画像を呼び出して、Excelに貼り付け
    """


# ワークブック設定
wb = openpyxl.Workbook()
ws = wb.worksheets[0] # 1番目のシートを編集対象にする
ws.title = SHEET_TITLE # 1番目のシートに名前を設定
files = glob.glob('*.jpg')
print(files)
data = {'a':[files[0], files[1]]}

attach_img(data, ws)

# ファイルへの書き込み
wb.save(RESULT_FILE_NAME)
```
